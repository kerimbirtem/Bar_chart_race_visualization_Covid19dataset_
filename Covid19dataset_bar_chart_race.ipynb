```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('corona_dat.csv')
```

```python
pd.set_option('display.max_columns',None) # df de tüm kolonları göstermesini sağlar
df.head()
```

```python
df.info()
```

```python
#Türkiyeyi ayrı bir seriye at
df_tr=df['Turkey']
type(df_tr)
```

```python
#türkiyedeki günlük vaka değişimini çiz
df_tr.plot()
```

```python
#Seçilen ülkelerdeki günlük vaka sayısı değişimini çiz
df_countries=df[['Turkey','Italy','Germany','China','US']]
df_countries.plot()
```

```python
#seçilen ülkelerdeki toplam vaka sayısı
df_countries.sum(axis=0)
```

```python
#Toplam vaka sayısı
df_countries.sum(axis=0).plot(kind='bar')
```

```python
sum_daily=df.cumsum(axis=0) # aynı sütundaki rakamları alt alta toplasın
sum_daily
```

```python
sum_daily['Countries_Total']=sum_daily.sum(axis=1)
sum_daily
```

```python
sum_daily.plot()
```

```python
#tarihsel olarak sıralı olduğu için date ı index oalrak ayarlayabiliriz
df['date']=pd.to_datetime(df['date'])
df.set_index('date',inplace=True) #indexi i date olarak ayarla tarihe göre index le
df
```

```python
#vaka sayıalrını kümülatif toplam yap
df.cumsum(axis=0).plot()
```

```python
df_yarisan=df.cumsum(axis=0) # Günlğük vakaların kümülatif toplamlarında bir veri seti oluştur
df_yarisan
```

```python
sns.lineplot(data=df_yarisan)
```
